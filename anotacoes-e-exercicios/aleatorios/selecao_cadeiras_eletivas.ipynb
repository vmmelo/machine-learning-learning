```python
import pandas as pd
eletivas_cc = pd.read_csv('inputs/eletivas-cc.csv')
eletivas_periodo_atual = pd.read_csv('inputs/eletivas-2020-1.csv')
eletivas_cc.tail()
```

```python
import re
def get_codigo_disciplina_cc(nome):
    search = re.findall("[a-zA-Z]{2}\d+", nome) if isinstance(nome, str) else []
    return (len(search) > 0 and search[len(search) - 1]) or ''

# separa codigo da disciplina
eletivas_cc['cod_disciplina'] = eletivas_cc.apply(lambda row: get_codigo_disciplina_cc(row['disciplina']), axis=1)
eletivas_cc.tail()
```

```python
def get_codigo_disciplina_periodo_atual(nome):
    search = re.findall("[a-zA-Z]{2}\d+", nome) if isinstance(nome, str) else []
    return (len(search) > 0 and search[len(search) - 1]) or ''

# separa codigo da disciplina
eletivas_periodo_atual['cod_disciplina'] = eletivas_periodo_atual.apply(lambda row: get_codigo_disciplina_periodo_atual(row['disciplina']), axis=1)
eletivas_periodo_atual.tail()
```

```python
eletivas_pagas = eletivas_cc[eletivas_cc['nota'] >= 7]
eletivas_pagas = eletivas_pagas['cod_disciplina'].to_numpy()
eletivas_pagas
```

```python
eletivas_periodo_atual['pago'] = eletivas_periodo_atual.apply(lambda row: 1 if row['cod_disciplina'] in eletivas_pagas else 0, axis=1)
eletivas_periodo_atual.tail()
```

```python
eletivas_a_pagar = eletivas_periodo_atual[eletivas_periodo_atual['pago'] == 0]
remove_professores = ['Ruy Guerra']
eletivas_a_pagar = eletivas_periodo_atual[~eletivas_periodo_atual['professor'].isin(remove_professores)]
eletivas_a_pagar
```
